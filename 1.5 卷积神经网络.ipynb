```python
#使用卷积神经网络实现手写数字识别
import tensorflow as tf
import numpy as np
from tensorflow.examples.tutorials.mnist import input_data

#数据集
minst = input_data.read_data_sets('MNIST_data', one_hot = True)

#定义两个占位符
x = tf.placeholder(tf.float32, [None, 784])
y = tf.placeholder(tf.float32, [None, 10])
# k_prob = tf.placeholder(tf.float32, [None, 30])
#定义一个batchSize
batchSize = 100
batchNum = minst.train.num_examples // batchSize

#将图像数据处理成CNN可以处理的结构
input_x = tf.reshape(x, [-1, 28, 28, 1])


#定义卷基层
conv1 = tf.layers.conv2d(
    inputs= input_x,
    filters = 32, #卷积核的厚度
    kernel_size = [5,5],
    strides=1, #扫描步长
    padding='same'
)
#池化层
pool1 = tf.layers.max_pooling2d(
    inputs= conv1,
    pool_size = [1, 1],
    strides = 2
)
#卷积层2
conv2 = tf.layers.conv2d(
    inputs=pool1,
    filters=64,
    kernel_size=[5,5],
    strides=1,
    padding='same',
    activation=tf.nn.relu
)
#池化层2
pool2 = tf.layers.max_pooling2d(
    inputs=conv2,
    pool_size=[2,2],
    strides=2
)
print(pool2)

# #将最后的结果平坦化
flat = tf.reshape(pool2, [-1, 7 * 7 *64])

#链接全链接层
dense = tf.layers.dense(
    inputs=flat,
    units=1024,
    activation=tf.nn.relu
)

#可以选择丢弃部分的神经元防止过拟合
droupout = tf.layers.dropout(
    inputs = dense,
    rate = 0.5
)

#最后一层输出层
logits = tf.layers.dense(
    inputs = droupout,
    units = 10
)

#定义损失函数
loss = tf.losses.softmax_cross_entropy(onehot_labels=y, logits=logits)

#使用梯度下降法训练损失函数
train_op = tf.train.GradientDescentOptimizer(learning_rate=0.001).minimize(loss)

#计算准确率
correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(logits, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))


init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    for epoh in range(1):
        for nextbatch in range(batchNum):
            batchX, batchY = minst.train.next_batch(nextbatch)
            sess.run([loss, train_op], feed_dict={x:batchX, y:batchY})
            print(sess.run(accuracy,feed_dict={x:batchX, y:batchY}))
        
```
